1.¿Que tipo de organizacion cuenta con mas matriculas activas?

```python
razonSocial = dx[['ORGANIZACION', 'EST-MATRICULA']]
razonSocial.shape
```

```python
alt.Chart(razonSocial).mark_bar().encode(    # Making a bar chart
    alt.X('count():Q'),                    # X-axis is an aggregated value for categorical variable in Y
    alt.Y('ORGANIZACION:N', sort = '-x'),   # Y-axis is sorted based on aggregated value
    alt.Color('EST-MATRICULA:N'),
    tooltip = [alt.Tooltip('count():Q')]
).properties(
    width = 600,
    height = 150
)
```

2. Ultimo año renovado con el tipo de organizacion

```python
lastRe = dx[['ORGANIZACION', 'ULT-ANO_REN']]
lastRe.shape
```

```python
alt.Chart(lastRe).mark_bar().encode(    # Making a bar chart
    alt.X('count():Q'),                    # X-axis is an aggregated value for categorical variable in Y
    alt.Y('ORGANIZACION:N', sort = '-x'),   # Y-axis is sorted based on aggregated value
    alt.Color('ULT-ANO_REN:N'),
    tooltip = [alt.Tooltip('count():Q')]
).properties(
    width = 600,
    height = 150
)
```

En la anterior grafica se puede filtrar la información de la mayoria de organizaciones cuentan con su renovacion al dia, es decir la cantidad de empresas que deben actualizar son muy pocas comparado con las empresas ya totalmente renovadas

```python
tipoDeEmpresa = dx[['CATEGORIA','ORGANIZACION']]
tipoDeEmpresa.shape
```

```python
alt.Chart(tipoDeEmpresa).mark_point(filled = True).encode(
    alt.X('ORGANIZACION'),
    alt.Y('CATEGORIA'),
    alt.Size('count():Q'),
    alt.Color('ORGANIZACION'),    # Using the color channel for encoding a categorical variable
    alt.OpacityValue(3)     # Adding size as channel for encoding a third quantitative variable
).properties(
    width = 600,
    height = 500
)
```

Que cantidad de categorias hay por tipo de organizacion?
Se pudo dilucidar de la informacion recolectada que se encuentra un tipo de empresa conocido como establemciento de comercio que no tienen nigun tipo de categoria, que no aplican porque suelen ser tipos de tiendas normales, no son una agencia de una empresa mas grande, son empresas que tienen una sola sede, se decidio utilizar este tipo de diagrama por su vista mas didactica frente a la cantidad de datos ya que tenemos referencias de que cantidad hay en base al tamaño del punto

```python
municipo = dx[['MUN-NOTIFICACION', 'MUN-COMERCIAL']]
```

```python
alt.Chart(municipo)\
    .mark_point().encode(            # Defining point as mark
        alt.X('count()'),  # Definig position in X as channel for encoding a quantitative variable
        alt.Y('MUN-NOTIFICACION'),
        alt.Size('count()'),
    alt.Color('MUN-NOTIFICACION'),    # Using the color channel for encoding a categorical variable
    alt.OpacityValue(2)     # Definig position in Y as channel for encoding other quantitative variable
    ).properties(
        width = 600,
        height = 500
    )
```

¿Cuantas empresas estan registradas fuera de villavicencio?
Se puede dar cuenta que aproximadamente el 39% de empresas no tienen sede principal en villavicencio, solo tienen una sede y/o agencia en villavicencio, se utilizo este diagrama de puntos donde se le asigno a colores diferentes a cada municipio para poder diferenciar de manera correcta cada uno.
